The following function comes from the central-limit-theorem-assignment.md file

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def make_draws(dist, params, size=200):
    """
    Draw a sample from a specified distribution
    with given parameters and return these in an array.

    Parameters
    ----------

    dist: scipy.stats distribution object:
      Distribution object from scipy.stats, must have a .rvs method

    params: dict:
      Parameters needed to define the distribution dist.
    For example, if dist = scipy.stats.binom, then params could be

          {'n': 100, 'p': 0.25}

    size: int:
      The number of values to draw

    Returns
    -------
    sample: np.array, shape (size, )
      An i.i.d sample from the specified distribution.
    """
    return dist(**params).rvs(size)

# Generate draws from the Binomial Distribution, using Scipy's binom object.
binomial_samp = make_draws(stats.binom, {'n': 100, 'p':0.25}, size=200)

plt.scatter(range(len(binomial_samp)),binomial_samp);
plt.title('200 values from a Binomial Distribution', fontsize=20)
```

```python
def plot_means(dist, params, ax, repeat, size=100):
    
    sample_means = [dist(**params).rvs(size).mean() for i in range(repeat)]
    return ax.hist(sample_means,bins=100)
```

### Binomial Distribution

```python
fig, ax = plt.subplots()
plot_means(stats.binom, {'n':100,'p':0.4}, ax, repeat=5000);
```

### Poisson Distribution

```python
fig, ax = plt.subplots()
plot_means(stats.poisson,{'mu':5}, ax, repeat=30000);
```

### Exponential Distribution

```python
fig, ax = plt.subplots()
plot_means(stats.expon,{'scale':1}, ax, repeat=5000);
```

### Geometric Distribution

```python
fig, ax = plt.subplots()
plot_means(stats.geom,{'p':0.3}, ax, repeat=20000);
```

### Uniform Distribution

```python
fig, ax = plt.subplots()
plot_means(stats.uniform,{'loc':0,'scale':100}, ax, repeat=5000);
```

4. What do you observe if you change the sample size to 10 instead of 200, keeping repeat constant at 5000? Explain your observation on a high level. Should the CLT apply when your sample size is small with a large value for repeat? Hint: this is best answered by distinguishing between the role of the of the size and the repeat variable.

I'll use the uniform distribution to answer this question since it was the one who looks more similar to a normal distribution with the lowest value of repeats.

```python
fig, ax = plt.subplots(2,3, figsize=(16,12))

sizes = [100,20,10,5,2,1]

for size_test, ax in zip(sizes, ax.flatten()):
    plot_means(stats.uniform,{'loc':0,'scale':100}, ax, repeat=5000, size = size_test);
    ax.set_xlim([0,100])
```

From the graph above we can see that as long as the size of the samples are > 1 the CLT applies. Decreasing the size increases the variance of the data, which makes sense because with higher sample size the mean will be closer to the population mean.

When the size is equal to 1 it's not possible to see the effect since this means just getting 1 value from the distribution not an actual sample mean.

5. Instead of taking the mean of the samples, take the maximum of each of the samples and plot the histograms again. Do they resemble the normal distribution? Do all sample statistics follow a normal distribution? In your answer clarify the sample statistics to which the CLT applies.

```python
def plot_max(dist, params, ax, repeat, size=100):
    
    sample_max = [dist(**params).rvs(size).max() for i in range(repeat)]
    return ax.hist(sample_max,bins=100)

def plot_min(dist, params, ax, repeat, size=100):
    
    sample_min = [dist(**params).rvs(size).min() for i in range(repeat)]
    return ax.hist(sample_min,bins=100)

def plot_std(dist, params, ax, repeat, size=100):
    
    sample_std = [dist(**params).rvs(size).std() for i in range(repeat)]
    return ax.hist(sample_std,bins=100)
```

```python
fig, ax = plt.subplots(1,3, figsize=(16,12))

for i, ax in zip(range(3), ax.flatten()):
    
    if i == 0:
        plot_max(stats.uniform,{'loc':0,'scale':100}, ax, repeat=5000)

    elif i == 1:    
        plot_min(stats.uniform,{'loc':0,'scale':100}, ax, repeat=5000)

    else:
        
        plot_std(stats.uniform,{'loc':0,'scale':100}, ax, repeat=5000)
```

We can see that (besides the mean) the other statisitic that also follows the central limit theorem is the standard deviation. This makes sense since these two parameters are the parameters for a normal distribution. In this case the standard deviation of a uniform distribution is:

$$ \sigma^2 = \frac{(a-b)^2}{12}$$

Using the previous formula for our case:

```python
sigma = np.sqrt(((0-100)**2)/12)
sigma
```

So we can conclude that the sample std tends to be the population's std.

2. Viewing the lunch our data as an i.i.d sample from the lunch hours of ALL Google employees, the sample mean is a random variable, and we have a single draw from it. What is the distribution of this sample mean random variable, and how do you know?

To answer this question the CLT will be used fit a model.

```python
df_mean = df.mean()[0]
df_std = df.std()[0]/np.sqrt(len(df[0]))

print('Lunch time mean: ', round(df_mean,3))
print('Lunch time std: ', round(df_std,3))
```

Therefore the sample means follows a normal distribution with a mean of 2.2 and a std of 0.041 (also known as standard error).

3. Based on the standard error and the sample mean, compute the 95% confidence interval for the population mean lunch break length.

```python
upper = df_mean + (1.96 * df_std)
lower = df_mean - (1.96 * df_std)

print('Upper bound: ',round(upper,2))
print('Lower bound: ',round(lower,2))

```

```python
df['center'] = df[0] - df_mean
dist = stats.norm(df['center'].mean(), df_std)
upper = dist.ppf(0.975)
lower = dist.ppf(0.025)
print(upper)
print(lower)
print(df_mean + upper)
print(df_mean - upper)
```
