## Part 1

### Scatterplot with conditional color

* Create two arrays of random integers.
* Create a third array that evaluates if the sum of the integers of a particular point is even. This array will be used to determine the colot of the points in the scatter plot.

```python
a = np.random.randint(1000, size=50)
b = np.random.randint(1000, size=50)
c = ['r' if  (x+y)%2 == 0 else 'b' for x,y in zip(a,b)]
```

```python
fig, ax = plt.subplots()

ax.scatter(a, b, c = c)
```

### Two functions

```python
linear = lambda x: 3*x + 0.5
sqr = lambda x: 5*x**(0.5)
x = np.linspace(0,10,51)

y_1 = linear(x)
y_2 = sqr(x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, y_1,linestyle=':', c = 'r')
ax.plot(x,y_2,linestyle='-', c ='b')
ax.legend(["Linear","Sqrt"])

# Graph with log scale
#ax.set_xscale('log')
```

```python
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
fig, ax = plt.subplots()
ax.bar(np.arange(len(barheights)), barheights)
x_pos = np.arange(len(barheights))
ax.set_xticks(x_pos)
ax.set_xticklabels(barlabels, rotation=45)
```

* To change the x position of the labels we can pass a list as an argument on set_xticklabels with the order we want from the dataset.

```python
barheights = [3,5,1]
barlabels = ['grapes', 'oranges', 'hockey pucks']
fig, ax = plt.subplots()
ax.bar(np.arange(len(barheights)), barheights)
x_pos = np.array([0,1,2])
ax.set_xticks(x_pos)
ax.set_xticklabels(['oranges','grapes','hockey pucks'], rotation=45)
```

* So the answer here is that, changing x_pos changes how the labels on set_sticklabels will be displayed.
